```python
import pandas as pd

df = pd.read_csv("healthcare_data.csv")
correlation = df['Blood_Pressure'].corr(df['Cholesterol_Level'])
print("Correlation:", correlation)
```

```python
mean_cost = df.groupby('Smoker')['Treatment_Cost'].mean()
print(mean_cost)
```

```python
df['Age_Group'] = pd.cut(df['Age'], bins=[0,18,35,50,65,100],
labels=['0-18','19-35','36-50','51-65','65+'])

visits = df.groupby('Age_Group')['Hospital_Visits'].sum()
print(visits)
```

```python
bmi_variance = df['BMI'].var()
bmi_std = df['BMI'].std()

print("Variance:", bmi_variance)
print("Standard Deviation:", bmi_std)
```

```python
diabetic_percentage = (df['Diabetes'].value_counts(normalize=True) * 100)
print(diabetic_percentage)
```
