```python
import pandas as pd
df = pd.read_csv("healthcare_data.csv")
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=10)

plt.title("Distribution of Patient Ages")
plt.xlabel("Age")
plt.ylabel("Number of Patients")

plt.show()
```

```python
df['Gender'].value_counts().plot(kind='bar')

plt.title("Gender Distribution of Patients")
plt.xlabel("Gender")
plt.ylabel("Number of Patients")

plt.show()
```

```python
plt.scatter(df['BMI'], df['Treatment_Cost'])

plt.title("BMI vs Treatment Cost")
plt.xlabel("BMI")
plt.ylabel("Treatment Cost")

plt.show()
```

```python
df['Age_Group'] = pd.cut(df['Age'],
                         bins=[0,18,35,50,65,100],
                         labels=['0-18','19-35','36-50','51-65','65+'])

avg_visits = df.groupby('Age_Group')['Hospital_Visits'].mean()

avg_visits.plot(kind='line', marker='o')

plt.title("Average Hospital Visits by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Hospital Visits")

plt.show()
```

```python
df['Smoker'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Smoker vs Non-Smoker Distribution")
plt.ylabel("")

plt.show()
```
